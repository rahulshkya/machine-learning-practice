```python
import requests
import pandas as pd
URL="https://www.scrapethissite.com/pages/simple/"
res=requests.get(URL)
html=res.text
with open("Scraped_data/p.html", "w", encoding="utf-8") as f:
    f.write(html)
```

```python
#beautifulSoup
from bs4 import BeautifulSoup
with open("Scraped_data/p.html","r") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")
```

```python
all_h3=soup.find_all("h3")
all_countries=[]
for h3 in all_h3:
    name=h3.get_text(strip=True) 
    # print(h3.find_parent("div")["class"])
    population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    all_countries.append([name,population])
```

```python
import pandas as pd
df=pd.DataFrame(all_countries,columns=["Name","population"])
df.to_csv("clean_data/lean.csv")
```
