```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
        
    return data
```

```python
data=load_data("store_data.json")
```

```python
print(data)
print(type(data))
```

```python
#clean and structure the data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    
    for user in data:
        #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating

        #handling missing value
        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None

        #deduplication
        
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)
```

```python
#get meaningful data
def get_insights(data):
    #get average rating
    tot_rating=0
    poor_rating=0
    for user in data:
        tot_rating = tot_rating + float(user["rating"])
   
    print(f"avg  rating = {tot_rating/len(data)}")

    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating = poor_rating + 1
    print(f"poor rating average ={poor_rating/len(data)}")
```

```python
get_insights(data)
```

```python
def get_rec(data):
    reccommed_data=[]
   
     
    for user in data:
        curr_recc ={}
        curr_recc["name"]=user["name"]
        if(float(user["rating"]) >= 3 ):
            curr_recc["brand"]="Apple"
        else:
            curr_recc["brand"]="Samsung"

        reccommed_data.append(curr_recc)
    return reccommed_data
```

```python
get_rec(data)
```
