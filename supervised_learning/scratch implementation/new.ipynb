```python

import pickle

model = pickle.load(open("model.pkl","rb"))
scaler = pickle.load(open("scaler.pkl","rb"))
```

```python
import pandas as pd

test1 = pd.DataFrame(
    [[7,6,3,80,4,1]],
    columns=["study","sleep","screen","attendance","stress","exercise"]
)

test2 = pd.DataFrame(
    [[2,4,8,60,9,0]],
    columns=["study","sleep","screen","attendance","stress","exercise"]
)

print(model.predict(scaler.transform(test1)))
print(model.predict(scaler.transform(test2)))
```
