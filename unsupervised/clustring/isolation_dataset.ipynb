```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("thyroid_dataset.csv")
```

```python
df.head()
```

```python
X=df.drop("Outlier_label",axis=1)
y=df["Outlier_label"]
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.ensemble import IsolationForest
clf=IsolationForest(n_estimators=200,contamination="auto",random_state=42)
```

```python
labels=clf.fit_predict(X_scaled)
```

```python
labels
```

```python
#visualize
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
```

```python
import numpy as np
n_out=np.sum(labels == -1)
n_normal=np.sum(labels == 1)

print("outliers : ",n_out)
print("n_normal : ",n_normal)
```

```python
#LOF
from sklearn.neighbors import LocalOutlierFactor
```

```python
neighbs=LocalOutlierFactor(contamination=0.36)
labels=neighbs.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
```
