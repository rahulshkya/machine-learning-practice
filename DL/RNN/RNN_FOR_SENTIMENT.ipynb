## Preprocessing

# 1.Convverting to lowercase

```python
df["review"]=df["review"].str.lower()
```

# 2.remove links

```python
import re
# sample_text="abc is the word,abc" #abc => xyz

# new_text=re.sub("abc","xyz",sample_text)
```

```python
import re

def remove_urls(text):
    return re.sub(r"http\S+", "", str(text))

df["review"] = df["review"].apply(remove_urls)
```

```python
df.head
```

# 3. removing punctuation

```python

def remove_punc(text):
    if pd.isna(text):
        return text
    return re.sub(r"[^A-Za-z0-9\s]", "", str(text))

df["review"] = df["review"].apply(remove_punc)
```

# 4.Removing html

```python
def remove_html(text):
    if pd.isna(text):
        return text
    return re.sub(r"<.*?>", "", str(text)) #. = all, * = 0-10 , ? =non greedyy

df["review"] = df["review"].apply(remove_html)
```

# 5.Reemoving the StopWords

```python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd

stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
    if pd.isna(text):
        return ""

    tokens = word_tokenize(str(text))

    filtered_words = [word for word in tokens if word.lower() not in stop_words]

    return " ".join(filtered_words)

df["review"] = df["review"].apply(remove_stopwords)
```

# 6.stemming

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemming(text):
    ps = PorterStemmer()
    stemmed_words = []

    tokens = word_tokenize(str(text))

    for word in tokens:
        stemmed_token = ps.stem(word)
        stemmed_words.append(stemmed_token)

    return " ".join(stemmed_words)

df["review"] = df["review"].apply(stemming)
```

```python

type(X_test)
X_test=X_test.toarray()
X_train=X_train.toarray()

# numpy.ndarray
```

```python

train_set=TensorDataset(
    torch.from_numpy(X_train).float(),
    torch.from_numpy(y_train.values).float()
)

test_set =TensorDataset(
    torch.from_numpy(X_test).float(),
    torch.from_numpy(y_test.values).float()
)
```

```python
train_loader=DataLoader(train_set,shuffle=True,batch_size=64)
test_loader=DataLoader(test_set,shuffle=True,batch_size=64)
```

# Build our RNN

```python
import torch.nn as nn
import torch.optim as optimizer
```

```python
class RNN(nn.Module):
    def __init__(self,input_size,hidden_size=128,num_layers=1):
        super().__init__()

        self.hidden_size=hidden_size
        self.num_layers=num_layers

        #RNN layer
        self.rnn=nn.RNN(input_size,hidden_size,num_layers,batch_first=True)

        #fully connected layer
        self.fc = nn.Linear(hidden_size,1)
    def forward(self,x):
        #optional
        h0= torch.zeros(self.num_layers,x.size(0),self.hidden_size)

        out, _=self.rnn(x,h0)

        out=self.fc(out[:,-1,:])

        return out
```

```python
import torch.optim as optim

input_size = X_train.shape[1]

model=RNN(input_size)

criterion=nn.BCELoss()
optimizer=optim.Adam(model.parameters())
```

```python
# train the rnn
```

```python
epochs=10

for epoch in range(epochs):
    model.train()

    for Xb,yb in train_loader:
        optimizer.zero_grad()

        Xb=Xb.unsqueeze(1)
        outputs=model(Xb) #2D

        outputs=torch.sigmoid(outputs.squeeze()) #1D
        
        loss=criterion(outputs,yb) #compute loss
        loss.backward()
        optimizer.step()

    print(f"{epoch}/{epochs} and loss = {loss.item()}")
```

```python
#evaluate
  
model.eval()

with torch.no_grad():
    correct_vals=0
    tot_vals=0 

    for Xb,yb in test_loader:
        Xb =Xb.unsqueeze(1) 

        outputs=model(Xb)
        predicted=(torch.sigmoid(outputs.squeeze()) > 0.5).float()

        tot_vals += yb.size(0)
        correct_vals+=(predicted == yb).sum().item()

    print(f"accuracy ={correct_vals/tot_vals*100}")
```
