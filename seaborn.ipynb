```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="white")
sns.get_dataset_names()
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
x_value=[1,2,3,4,5,6,7,8]
y_value=[i**2 for i in x_value]

sns.relplot(
    x=x_value,
    y=y_value,
    kind="scatter"
)
```

```python
penguins=sns.load_dataset("penguins")
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
penguins.head()
```
