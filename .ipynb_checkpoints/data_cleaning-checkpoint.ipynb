```python
import pandas as pd
```

```python
df=pd.read_csv("raw_data.csv")
df2=df.copy()
df2["tax"]=df2["income"].apply(lambda x : "20%" if x >= 50000 else "10%")
gender={
    "Male":"M",
    "Female":"F",
    "Unknown":"U"
}

df2["gender"].map(gender)
df2.assign(new_income = df2["income"] * 1.1 ) 
```
